The youtuber uses the Python Imaging Library (PIL) to create and image and draw on the image

To use the PIL module you first need to "import" it

```python
import random
from PIL import Image, ImageDraw, ImageShow
```

first part of program sets up the image size
It also sets the size of the squares that will be drawn on the image
It also sets how many squares will be drawn

```python
image = Image.new("RGB", (1000, 1000))
width, height = image.size

# sets the height and width of the squares
rectangle_width = 50
rectangle_height = 50

# this line randomly decides how many squares will be show
# but the random number is between 10 and 550
number_of_squares = random.randint(10, 550)
```

Next part (1) randomly selects a position on image (2) sets top-left plus bottom-right points of rectangle 
(3) draws rectangle plus fills it using a random colour

In PIL the top left hand corner of the image equals 0,0

```python

draw_image = ImageDraw.Draw(image)

for i in range(number_of_squares):
    rectangle_x = random.randint(0, width)
    rectangle_y = random.randint(0, height)
    
    rectangle_shape = [
        (rectangle_x, rectangle_y),
        (rectangle_x + rectangle_width, rectangle_y + rectangle_height)]
    draw_image.rectangle(
        rectangle_shape,
        fill=(
            random.randint(0, 255), 
            random.randint(0, 255), 
            random.randint(0, 255)
        )
    )
```
